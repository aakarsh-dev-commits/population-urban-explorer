```python
world["area_deg2"] = world.geometry.area
world[["ADMIN" , "area_deg2"]].head()
```

```python
world_eq = world.to_crs("EPSG:6933")
world_eq.crs
```

```python
world_eq["area_km2"] = world_eq.geometry.area / 1e6
world_eq[["ADMIN" , "area_km2"]].head()
```

```python
pop = pd.read_csv(POP_DATA)
```

```python
pop.columns
```

```python
pop.head()
```

```python
world_pop = world_eq.merge(
    pop,
    how = "left",
    left_on = "ADM0_A3",
    right_on = "Country Code"
)
```

```python
world_pop["2022 [YR2022]"].isna().sum()
```

```python
missing = world_pop[world_pop["2022 [YR2022]"].isna()]
```

```python
world_pop["2022 [YR2022]"].dtype
```

```python
world_pop["population"] = pd.to_numeric(
    world_pop["2022 [YR2022]"],
    errors = "coerce"
)
```

```python
world_pop["population"].dtype
```

```python
world_pop["population"].isna().sum()
```

```python
world_pop["pop_density"] = (
    world_pop["population"] / world_pop["area_km2"]
)
```

```python
world_pop.sort_values("pop_density" , ascending = False)[["ADMIN","pop_density"]].head(10)
```

```python
world_pop["log_density"] = np.log10(world_pop["pop_density"])
```

```python
world_pop[["ADMIN" , "log_density" , "pop_density"]].head()
```

```python
world_pop["log_density"].min(), world_pop["log_density"].max()
```

```python
log_ticks = np.array([-1,0,1,2,3])
```

```python
fig , ax = plt.subplots(1,1, figsize=(15,18))

world_pop.plot(
    column = "log_density",
    cmap = "viridis",
    linewidth=0.2,
    ax = ax,
    missing_kwds={
        "color": "lightgrey",
        "label": "No data"
    },
    legend=True
)

cbar = ax.get_figure().axes[-1]

cbar.yaxis.set_ticks(log_ticks)
cbar.yaxis.set_ticklabels([f"{np.power(10.0, t):g}" for t in log_ticks])


ax.set_title(
    "Global Population Density (log-scaled, people per km2)",
    fontsize=14
)

ax.axis("off")

plt.show()
```
